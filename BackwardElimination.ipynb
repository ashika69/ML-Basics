We can measure regression by 4 methods ,

**MEAN SQUARE ERROR** ---> (y_actual - y_predicted)^2 / n

 **ROOT MEAN SQUARE ERROR** ---> similar to mean square error
                  
                  sqrt((y_actual - y_predicted)^2)/n))
**R - SQUARED ** ---> Mostly used to know the correlation between the variables
   
   --> Here , the value increases if it's has more correlation otherwise it stays constant and doesn't  decresae.

   -->Hence, it creates a ambuigty whether it's correlated or not

**ADJUSTED R-SQUARED ** ---->Same as of r-square , here the value decreases if it is correlation is low.

Here , we can go **BACKWARD ELIMINATION** , and delete those independent variables thus,may increase the accurancy.


   

```python
import numpy as np
import pandas as pd
```

```python
dataset= pd.read_csv('50_Startups.csv')
```

```python
dataset.head()
```

```python
x = dataset.iloc[: , :-1].values
y = dataset.iloc[: , 4].values
```

Here , our dataset consists of **Categorial Data**--> states.

therefore , by using

 **LabelEncoder , ColumnTransformer , OneHotEncoder** ---> we will encode the encoding of categorial data

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
labelencoder = LabelEncoder()
x[: , 3] = labelencoder.fit_transform(x[: , 3])
ct = ColumnTransformer([('encode' , OneHotEncoder() ,[3])], remainder='passthrough')
x = ct.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_tarin , y_test = train_test_split(x, y , test_size= 0.2 , random_state = 0)
```
