**Matplotlib**
A python lib used to plot graphs in different style like lines,bars,pie...

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([10,20,30,40])
y = np.array([10,20,30,40])
plt.plot(x,y)
plt.show()
```

To show dots/points we use keyword **marker** as a parameter and mention the type of points as dots,star,p...

marker = 'o'/'*'/'p'....

```python
plt.plot(x,y,marker='o')
plt.show()
```

We can add features as color,size,markeredge color by passing as parameters for color as c = 'color' , size as s = 'int size' , mec = 'color' and mfc = 'color'

```python
plt.plot(x,y,marker='*' , c = 'r' , ms = 10 , mec = 'k',mfc = 'b')
plt.show()
```

**Lines**

default it gives solid line ,we can change lines in matplotlib by using linestyle as ls = ':' (dotted) , '--'(dashed) ,'.-'(dotdashed) as parameters

```python
plt.plot(x,y, marker='o' , linestyle='-.' , c='r', lw=2 ,ms=10 , mfc = 'k' , mec = 'b')
plt.show()
```

Label and Title

```python
plt.xlabel("subject")
plt.ylabel("Time")
plt.plot(x,y)
plt.title("Ashika" , loc = 'left') #loc() is used to fix the position
plt.grid() #used to get grid lines ,
plt.show()
```

```python
plt.xlabel("subject")
plt.ylabel("Time")
plt.plot(x,y)
plt.title("Ashika" , loc = 'left') #loc() is used to fix the position
plt.grid(axis = 'x') #used to get grid lines , axis to x give y grid and vice versa
plt.show()
```

Scatter points - scatter(x,y) used to show scatter points

```python
plt.scatter(x,y)
plt.show()
```

we can add colors to points and we can get them by colormap as cmap

```python
plt.xlabel("subject")
plt.ylabel("Time")
mysize = np.array([30,60,90,120])
c = np.array(['red','green','blue' , 'black'])
plt.scatter(x,y , s=mysize , alpha = 0.5 , color = c , cmap = 'viridis' )
plt.colorbar()
plt.title("Ashika" , loc = 'left') #loc() is used to fix the position
```

Bar graphs by bar() method and horizontal bars by barh()

```python
plt.bar(x,y , color = 'red' , width = 5)
plt.show()
```

```python
plt.barh(x,y)
plt.show()
```

Histogram - used for frequency data and data should be normailzed(68 95 98) Rule

hist() - used to create histogram.

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

**Pie** - Pie() is used to create pie




shadow = True

labels = list of labels

explode= wedge out

legend() = used for list of edges

```python
x = [30,40,50]
lab = ['c','c++','python']
plt.pie(x , labels = lab , explode = [0.2 , 0 ,0] , shadow =  True)
plt.legend()
plt.show()
```

**NORMAL DISRTIBUTION**

The data is normally distributed follows the (68-95-98) rule , where the mean will be **zero** and std dev is **one** .

Numpy as **random.noraml()** method that takes mean,std dev and no. of values for the distribution - used to create big normal distributed data.

Histogram is used to represent the NDD. by hist() method and we can also pass no. of bars required as **plt.hist(x,100)**

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.normal(5.0, 1.0, 1000)
plt.hist(x,100)
plt.show()
```

```python
import numpy
import matplotlib.pyplot as plt

x = numpy.random.uniform(0.0, 5.0, 250)
plt.hist(x)
plt.show()
```
